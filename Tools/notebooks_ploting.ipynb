```python
#for bar graph
import matplotlib.pyplot as plt
import numpy as np
x_axis = ["16", "32", "64", "128", "400", "800"]
y_axis = [100,200,300,400,500,600]
width=0.6
plt.bar(x_axis, y_axis,width)
plt.title('fWave Solver with Intrinsics')
plt.xlabel('elements per dimension')
plt.ylabel('run-time(ms)')
plt.show()
```

```python
#for group bar graph
import matplotlib.pyplot as plt
import numpy as np


labels = ['16x16', '32x32', '64x64', '400x400', '800x800']
res_seq = [20, 34, 30, 35, 27]
res_SIMD = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  
width = 0.3  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, res_seq, width, label='without SIMD')
rects2 = ax.bar(x + width/2, res_SIMD, width, label='with SIMD')


ax.set_ylabel('Runtime')
ax.set_title('Vectorization')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```
